Линейная регрессия. Представлены данные, собранные путем голосования за самые популярные конфеты Хэллоуина. Необходимо обучить модель линейной многомерной регрессии. В качестве предикторов выступают поля: chocolate, fruity, caramel, peanutyalmondy, nougat, crispedricewafer, hard, bar, pluribus, sugarpercent, pricepercent, отклик — winpercent.В качестве тренировочного набора данных использовать данные из файла, за иключением следующих конфет: Haribo Gold Bears, Snickers.
Задания:
1.Вывести предсказанное значение winpercent для конфеты Haribo Gold Bears.
2.Вывести предсказанное значение winpercent для конфеты Snickers:
3.Вывести предсказанное значение winpercent для конфеты с параметрами:
[1, 0, 1, 1, 0, 1, 1, 1, 1, 0.355, 0.426]
4.Оценить точность модели.



```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
data = pd.read_csv("https://courses.openedu.ru/assets/courseware/v1/a11f0b9cb0074fdacc112eee71eb0b67/asset-v1:ITMOUniversity+DATANTECH2035+fall_2021_3+type@asset+block/candy-data.csv", delimiter = ',', index_col = 'competitorname')
data
```

Создадим тренировочный набор данных:исключив следующие конфеты: Haribo Gold Bears, Snickers.

```python
train_data = data.drop(['Haribo Gold Bears', 'Snickers'])
train_data
```

Отбор данных для предикторов:

```python
x = pd.DataFrame(train_data.drop(['winpercent', 'Y'], axis = 1))
x
```

Столбец отклика Y:

```python
y = pd.DataFrame(train_data['winpercent'])
y
```

Обучение модели:

```python
reg = LinearRegression().fit(x, y)
```

Предсказание для конфет, введенных в ручную:

```python
reg.predict([[1, 0, 1, 1, 0, 1, 1, 1, 1, 0.355, 0.426]])
```

Предсказание значения для конфеты Haribo Gold Bears:

```python
HariboGoldBears = data.loc['Haribo Gold Bears',:].to_frame().T
reg.predict(HariboGoldBears.drop(['winpercent','Y'], axis=1))
```

Предсказание значения для конфеты Snickers:

```python
Snickers = data.loc['Snickers',:].to_frame().T
reg.predict(Snickers.drop(['winpercent', 'Y'], axis=1))
```

Вывод коэффициентов θ0 и θ1,θ2 ... θp:

```python
reg.intercept_
```

```python
reg.coef_
```

Оценка точности модели:

```python
y_true = y['winpercent'].to_numpy()
y_new = reg.predict(x)
```

```python
r2_score(y_true,y_new)
```

Таким образом, точность предсказанной модели составляет 59%. Очевидно, что зависимость между признаками не является линейной.
